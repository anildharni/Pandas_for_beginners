```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
names = ['Anil','Vijaya','Shankaraiah','Swapna','Supps']
```

```python
dataset = [random.randint(0,1001) for i in range(0,1001)]
```

```python
bignames = [random.choice(names) for i in range(0,1001)]
```
